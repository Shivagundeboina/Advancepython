```python
import time
import sys
h=0
m=0
s=0
while True:
  sys.stdout.write("\r{:02d}:{:02d}:{:02d}".format(h,m,s))
  sys.stdout.flush
  time.sleep(1)
  s=s+1
  if s==60:
    s=0
    m=m+1
  if m==60:
    m=0
    h=h+1
  if h==13:
    h=1
  if h==0 and m==0 and s==0:
    h=0
print(h,m,s)

```
