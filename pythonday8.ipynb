```python
x="AAACCDDDAAD"
res=""
c=1
for i in range(len(x)):
  if i+1<len(x) and x[i]==x[i+1]:
    c=c+1
  else:
    res=res+x[i]
    res=res+str(c)
    c=1
print(res)
```

```python
class Rect:
  def __init__(self,a,b):
    self.area=a*b
  def display(self):
    print("area is ",self.area)

r=Rect(20,30)
r2=Rect(40,67)
r.display()
r2.display()
```
